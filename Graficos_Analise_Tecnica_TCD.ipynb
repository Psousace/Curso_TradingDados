##Osciladores


<p>
$ IFR = 100 - {\left(\frac{100}{1 + {\left(\frac{A}{B}\right)}}\right)} $

Onde:<p>
*A* = média dos preços dos últimos períodos de alta (usualmente fechamento e 14 períodos);<p>
*B* = média dos preços dos últimos períodos de baixa.<p>

```python
#Importar o módulo

from ta.momentum import RSIIndicator
```

```python
#Dataframe

#Criando o indicador
rsi = RSIIndicator(close=bragro['Close'], window=21)

#Trago para dentro do dataframe
bragro['RSI'] =  rsi.rsi()
```

```python
bragro
```

```python
#Gráfico de RSI

fig = make_subplots(rows=2, cols=1)
fig.add_trace(go.Scatter(name='RSI', x=bragro.index, y=bragro['RSI']),row=2, col=1)
fig.add_trace(go.Candlestick(name='BRAGRO', x=bragro.index, close=bragro['Close'], open=bragro['Open'], high=bragro['High'], low=bragro['Low']),row=1, col=1)
fig.update_layout(xaxis_rangeslider_visible=False)
fig.show()
```

###Oscilador Estocástico


- K = (P Fechamento - Min Mínima n períodos) / (Máx Alta - Mín Mínima) x 100
- D = média móvel de 3 períodos de K


```python
from ta.momentum import StochasticOscillator
```

```python
so = StochasticOscillator(high=bradesco['High'],
                          low=bradesco['Low'],
                          close=bradesco['Close'],
                          window=14)
```

```python
bradesco['SO_LINHA'] = so.stoch()
bradesco['SO_SINAL'] = so.stoch_signal()
```

```python
bradesco
```

```python
fig = make_subplots(rows=2,
                    cols=1,
                    row_heights=[3,0.7])
fig.add_trace(go.Scatter(name='K', x=bradesco.index, y=bradesco['SO_LINHA']), row=2,col=1)
fig.add_trace(go.Scatter(name='D', x=bradesco.index, y=bradesco['SO_SINAL']), row=2, col=1)
fig.add_trace(go.Candlestick(name='Bradesco', x=bradesco.index, close=bradesco['Close'], open=bradesco['Open'], high=bradesco['High'], low=bradesco['Low']), row=1, col=1)
fig.update_layout(xaxis_rangeslider_visible=False)
fig.show()
```

###Indicadores de Momentum

###Rate of Change

ROC = [(Close hoje - Close anterior de n períodos) / (Close anterior de n períodos)] x 100

```python
from ta.momentum import ROCIndicator
```

```python
roc = ROCIndicator(ouro['Close'])
ouro['ROC'] = roc.roc()
ouro
```

```python
fig = make_subplots(rows=2, cols=1,row_heights=[3,0.7])
fig.add_trace(go.Scatter(name='ROC', x= ouro.index, y=ouro['ROC']), row=2,col=1)
fig.add_trace(go.Candlestick(name='Ouro', x= ouro.index, open=ouro['Open'], close=ouro['Close'], high=ouro['High'], low=ouro['Low']), row=1, col=1)
fig.update_layout(xaxis_rangeslider_visible=False, title_text='ROC Ouro', template='plotly_dark')
fig.show()
```

###Percentage Price Oscillator (PPO)

PPO = [(EMA 12 - EMA 26)/ (EMA 26 )x 100]

```python
from ta.momentum import PercentagePriceOscillator
```

```python
PPO = PercentagePriceOscillator(ambev['Close'])
ambev['PPO'] = PPO.ppo()
ambev['PPO_hist'] = PPO.ppo_hist()
ambev['PPO_line'] = PPO.ppo_signal()
ambev
```

```python
fig = make_subplots(rows=2, cols=1, specs=[[{'secondary_y': True}], [{'secondary_y':True}]])
fig.add_trace(go.Scatter(name='PPO', x=ambev.index, y=ambev['PPO']), row=2, col=1)
fig.add_trace(go.Bar(name='PPO_Hist', x=ambev.index, y=ambev['PPO_hist']), row=2, col=1, secondary_y=True)
fig.add_trace(go.Scatter(name='PPO_Line', x=ambev.index, y=ambev['PPO_line']), row=2,col=1)
fig.add_trace(go.Candlestick(name='Ambev', x=ambev.index, close=ambev['Close'], open=ambev['Open'], high=ambev['High'], low=ambev['Low']))
fig.update_layout(xaxis_rangeslider_visible=False, title_text='PPO Ambev')
fig.show()
```

Commodity Chanel Index


CCI = (Preço Usual  -  MMS20p do Preço Usual) / (.015 x Desv Médio)

Preço Usual = (High + Low + Close)/3

Constante = .015

```python
from ta.trend import CCIIndicator
```

```python
cci_agro = CCIIndicator(high=bragro.High, low=bragro.Low, close=bragro.Close)
bragro['CCI'] = cci_agro.cci()
bragro
```

```python
fig = make_subplots(rows=2, cols=1, row_heights=[4,3])
fig.add_trace(go.Scatter(name='CCI', x=bragro.index, y=bragro['CCI']), row=2, col=1)
fig.add_trace(go.Candlestick(name='BRagro', x=bragro.index, close=bragro['Close'], open=bragro['Open'], high=bragro['High'], low=bragro['Low']))
fig.update_layout(xaxis_rangeslider_visible=False,
                  title_text='<b>CCI BRAGRO',
                  template='plotly_dark')
fig.show()
```

##MACD

```python
from ta.trend import MACD
```

```python
macd = MACD(bradesco.Close)
```

```python
bradesco['MACD_LINE'] = macd.macd() #Diferença entre a mm lenta e mm rapida
bradesco['MACD_SINAL'] = macd.macd_signal() #MM9 da MACD_LINE
bradesco['MACD_HIST'] = macd.macd_diff() #Histograma diferença entre macd e macd_signal
```

```python
bradesco
```

```python
fig = make_subplots(rows=2, cols=1)
fig.add_trace(go.Bar(name='MACD_Hist', x=bradesco.index, y=bradesco['MACD_HIST']), row=2,col=1)
fig.add_trace(go.Scatter(name='MACD_LINE', x=bradesco.index, y=bradesco['MACD_LINE']), row=2, col=1)
fig.add_trace(go.Scatter(name='MACD_SIGNAL', x=bradesco.index, y=bradesco['MACD_SINAL']), row=2, col=1)
fig.add_trace(go.Candlestick(name='BRADESCO', x=bradesco.index, close=bradesco['Close'],
                             open=bradesco['Open'], high=bradesco['High'],
                             low=bradesco['Low']))
fig.update_layout(xaxis_rangeslider_visible=False,
                  title_text='MACD Bradesco',
                  template='simple_white')
fig.show()
```
